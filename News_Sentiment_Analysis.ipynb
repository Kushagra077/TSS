```python
import requests
from bs4 import BeautifulSoup
from textblob import TextBlob
import newspaper
from googlesearch import search
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import WordNetLemmatizer
from newspaper import Article
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Define the query for Google Search
query = "mehul choksi india times"

# Use Google Search API to fetch the top 10 latest links for the query
links = []
for url in search(query, stop=10):
    links.append(url)

# Loop through each link and perform sentiment analysis
for link in links:
    # Use requests and BeautifulSoup to extract the text content of the article
    response = requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')
    text = soup.get_text()

    # Use TextBlob to perform sentiment analysis on the text
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity

    # Use newspaper to extract a summary of the article
    article = Article(link)
    article.download()
    article.parse()
    article.nlp()
    summary = article.summary
    title=article.title 

    # Print a summary of the sentiment analysis results and article summary for the current link
    print(f"Link: {link}")
    print("\n")
    print("Sentiment analysis result:")
    if polarity > 0:
        print("The news article has a positive sentiment.")
    elif polarity < 0:
        print("The news article has a negative sentiment.")
    else:
        print("The news article has a neutral sentiment.")

 
    #print the article title
    print(f"\nArticle:{title} ")

    # Print the article summary
    print(f"\nSummary: {summary}\n")
```

```python
import requests
from bs4 import BeautifulSoup
from textblob import TextBlob
from googlesearch import search

# Define the query for Google Search
query = "mehul choksi india times"

# Use Google Search API to fetch the top 30 links for the query
links = []
for url in search(query,stop=30):
    links.append(url)

# Loop through each link and perform sentiment analysis
for link in links:
    # Use requests and BeautifulSoup to extract the text content of the article
    response = requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')
    text = soup.get_text()

    # Use TextBlob to perform sentiment analysis on the text
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity

    # Determine the type of article based on the sentiment polarity
    article_type = ""
    if polarity > 0.2:
        article_type = "Positive"
    elif polarity < -0.2:
        article_type = "Negative"
    else:
        article_type = "Neutral"

    # Print the title, link, and sentiment analysis results for the current article
    print(f"Title: {soup.title}")
    print(f"Link: {link}")
    print(f"Article Type: {article_type}")
    print(f"Sentiment Polarity: {polarity}")
    print("----------------------\n")
```
